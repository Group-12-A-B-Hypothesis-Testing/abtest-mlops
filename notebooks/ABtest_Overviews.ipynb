Remove users from dataframe who do not respond to experiment

```python
def delete_users(data):
    cleand_df = data.query("not(yes==0 & no==0)")
    return cleand_df

clean_df = delete_users(df)
```

```python
#categorize the users as engaged or non-engaged
engaged = df[~((df['yes']==0) & (df['no']==0))]
notengaged = df[(df['yes']==0) & (df['no']==0)]
```
